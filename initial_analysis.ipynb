```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
```

```python
df_2024 = pd.read_csv('Data/play_by_play_2024.csv')
df_2024.head()
```

```python
fourth_down = df_2024[df_2024['down'] == 4.0]
fourth_down.head()
```

```python
fourth_down['play_type'].value_counts()
```

```python
fourth_down['go'] = (fourth_down['play_type'] == 'run') | (fourth_down['play_type'] == 'pass')
fourth_down['go'].value_counts()
```

```python
# plot distribution of going for it by distance
plt.figure(figsize=(10, 6))
sns.histplot(fourth_down[fourth_down['go'] == True]['ydstogo'])
plt.title('Distribution of Going for It by Distance')
plt.xlabel('Yards to Go')
```

```python
# correlation heat map for fourth down success
plt.figure(figsize=(10, 8))
subset = fourth_down[['fourth_down_failed', 'ydstogo', 'score_differential', 'half_seconds_remaining', 'game_seconds_remaining', 'go']]
corr = subset.corr()[['fourth_down_failed']]
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr.sort_values(by='fourth_down_failed'),
    annot=True,
    cmap='coolwarm',
    center=0
)
plt.title('Correlation with Fourth Down Failure')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
corr = subset.corr()[['go']]
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr.sort_values(by='go'),
    annot=True,
    cmap='coolwarm',
    center=0
)
plt.title('Correlation with Going for It')
plt.show()
```
